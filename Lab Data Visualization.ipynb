**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
query = """
SELECT CAST(CompTotal AS FLOAT) AS CompTotal
FROM main
WHERE CompTotal IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

plt.figure(figsize=(10,6))
plt.hist(df['CompTotal'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribuição de CompTotal')
plt.xlabel('CompTotal')
plt.ylabel('Frequência')
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
query = """
SELECT CAST(Age AS FLOAT) AS Age
FROM main
WHERE Age IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

plt.figure(figsize=(8,5))
plt.boxplot(df['Age'], vert=False)
plt.title('Boxplot da Idade')
plt.xlabel('Idade')
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
query = """
SELECT CAST(Age AS FLOAT) AS Age, 
       CAST(WorkExp AS FLOAT) AS WorkExp
FROM main
WHERE Age IS NOT NULL AND WorkExp IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

plt.figure(figsize=(8,6))
plt.scatter(df['Age'], df['WorkExp'], alpha=0.6)
plt.title('Relação entre Idade e Experiência de Trabalho')
plt.xlabel('Idade')
plt.ylabel('Experiência de Trabalho (anos)')
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
query = """
SELECT CAST(TimeSearching AS FLOAT) AS TimeSearching,
       CAST(Frustration AS FLOAT) AS Frustration,
       CAST(Age AS FLOAT) AS Age
FROM main
WHERE TimeSearching IS NOT NULL 
AND Frustration IS NOT NULL 
AND Age IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

plt.figure(figsize=(10,6))
plt.scatter(df['TimeSearching'], df['Frustration'], 
            s=df['Age']*5, alpha=0.5, color='blue')
plt.title('Bubble Plot: TimeSearching vs Frustration')
plt.xlabel('Tempo Pesquisando')
plt.ylabel('Frustração')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
query = """
SELECT DatabaseWantToWorkWith
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Explodir os bancos separados por ';'
db_list = df['DatabaseWantToWorkWith'].str.split(';').explode()

top5 = db_list.value_counts().nlargest(5)

plt.figure(figsize=(8,8))
plt.pie(top5, labels=top5.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Bancos de Dados que os Respondentes Querem Aprender')
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
query = """
SELECT CAST(Age AS INT) AS Age,
       CAST(TimeSearching AS FLOAT) AS TimeSearching,
       CAST(TimeAnswering AS FLOAT) AS TimeAnswering
FROM main
WHERE Age BETWEEN 30 AND 35
AND TimeSearching IS NOT NULL
AND TimeAnswering IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

grouped = df.groupby('Age').median()

plt.figure(figsize=(10,6))
plt.bar(grouped.index, grouped['TimeSearching'], label='TimeSearching')
plt.bar(grouped.index, grouped['TimeAnswering'], 
        bottom=grouped['TimeSearching'], label='TimeAnswering')

plt.title('Tempo Mediano Pesquisando e Respondendo (Idade 30-35)')
plt.xlabel('Idade')
plt.ylabel('Horas')
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
query = """
SELECT CAST(Age AS INT) AS Age,
       CAST(CompTotal AS FLOAT) AS CompTotal
FROM main
WHERE Age BETWEEN 45 AND 60
AND CompTotal IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

line_data = df.groupby('Age')['CompTotal'].median()

plt.figure(figsize=(10,6))
plt.plot(line_data.index, line_data.values, marker='o', color='green')
plt.title('CompTotal Mediano para Idades entre 45 e 60')
plt.xlabel('Idade')
plt.ylabel('CompTotal Mediano')
plt.grid(True)
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
query = """
SELECT MainBranch
FROM main
WHERE MainBranch IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

counts = df['MainBranch'].value_counts()

plt.figure(figsize=(8,6))
plt.barh(counts.index, counts.values, color='orange')
plt.title('Distribuição dos Respondentes por MainBranch')
plt.xlabel('Número de Respondentes')
plt.ylabel('MainBranch')
plt.show()
```
